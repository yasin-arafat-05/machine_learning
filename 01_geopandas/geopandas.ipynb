# `#1:Introduction: `

<br>

- GeoPandas is the open-source python package for reading, writing and analyzing the vector dataset.
- It extends the datatypes used by pandas to alow spatial operations on geometric types.
- It further depends on `fiona` for file access and matplotlib for visulization of data 

<br>

```python
import geopandas as gpd 
```

# `#Bangladesh Data:`

<br>
<br>

```python

bd = gpd.read_file("Bangladesh/upazilas.csv")
bd[:2]
```

```python
""" 
***********************************************************************
- Using Geopandas Convert -> longitute and latitude into Geometry POINT
***********************************************************************
"""

bd["geometry"] = gpd.points_from_xy(x=bd["lon"],y=bd["lat"])
```

```python
bd[:4]
```

```python

# Bangladesh shape file: 
bd_map = gpd.read_file("Bangladesh/bangladesh_geojson_adm0_whole_bangladesh_outline.json")
```

```python
bd_map[:4]
```

```python
bd_map.plot(edgecolor="black",color="green")
```

```python

#<----------------------------check the type of bd------------------->
# if pandas then we need to conver it gpd.geodataframe
print(type(bd))
print(type(bd_map))

print("CRS-> Coordinate Reference Should Be Same: ",bd_map.crs)
```

```python

# CRS-> should be same:
bd_gpd = gpd.GeoDataFrame(
    data=bd,
    geometry=gpd.points_from_xy(x=bd["lon"],y=bd["lat"]),
    crs="EPSG:4326"
)
```

```python
type(bd_gpd)
```

```python
bd_gpd.plot()
```

```python

#<============plot the data on bangladesh map===================>

fig,ax = plt.subplots(figsize=(10,4))
bd_map.plot(ax=ax,edgecolor='Black',color="green")
bd_gpd.plot(ax=ax)
plt.show()
```

```python
print(bd_map.head())
print(bd_map.crs)
print(bd_map.bounds)
```

```python
print(bd_gpd.total_bounds)  # [minx, miny, maxx, maxy]
print(bd_map.total_bounds)
```
