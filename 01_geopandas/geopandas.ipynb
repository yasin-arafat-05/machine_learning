# `#1:Introduction: `

<br>

- GeoPandas is the open-source python package for reading, writing and analyzing the vector dataset.
- It extends the datatypes used by pandas to alow spatial operations on geometric types.
- It further depends on `fiona` for file access and matplotlib for visulization of data 

<br>

```python

#%pip install geopandas 
```

```python
import geopandas as gpd 
```

```python

#load data: 
geo_json = r"shp/municipality.json"
shp = r"shp/municipality.shp"
```

```python

# read files:
gpddf1 = gpd.read_file(geo_json)
gpddf1.head(2)
```

```python

gpddf2 = gpd.read_file(shp)
gpddf2.head(2)
```

```python

# gdf coordinate system:
gpddf1.crs
```

```python

# geometry type:
gpddf1.geom_type
```

```python

#<-------------------This Nepal GeoGraphical Data------------------>
#visuliza the data:
gpddf1.plot()
```

```python

#<-------------------Seperate Data Base on Distict---------------------->
gpddf1.plot("DISTRICT",legend=True)
```

```python

import matplotlib.pyplot as plt 

fig,ax = plt.subplots(1,figsize=(10,4))
gpddf1.plot(ax=ax,column="DISTRICT",legend=True,legend_kwds={"loc":"center left"})
leg = ax.get_legend()
leg.set_bbox_to_anchor((1,0.5))
```
