# ordinal encoding & level encoding:

`We implement ordinal encoding in (review+education) and level encoding in purchased column.`

<br>
<br>


```python

df.drop(columns=['age','gender'],inplace=True)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
                                                df.drop(columns=['purchased']),
                                                df['purchased'],
                                                test_size=0.2
                                                        )

print(f"Size of X_train: {X_train.shape}")
print(f"Size of X_test: {X_test.shape}")
print(f"Size of y_train: {y_train.shape}")
print(f"Size of y_test: {y_test.shape}")
```

```python
X_train
```

```python

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

oe.fit(X_train)

X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```



#### `Here, out y contain categorical value. To convert this we need level encoder.`



```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(y_train)

le.classes_ # no: 0 ,yes: 1
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
