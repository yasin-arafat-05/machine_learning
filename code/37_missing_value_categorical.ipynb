# Missing value imputation: Categorical Value(Mode)

<br>
<br>

```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 

from sklearn.model_selection import train_test_split
```

```python

df = pd.read_csv("dataset/housing_train.csv",usecols=["GarageQual","FireplaceQu","SalePrice"])
df.head()
```

```python

df.info()
```

```python
df.isnull().mean() * 100
```

```python
sns.countplot(data=df,x="GarageQual")
```

```python
df["GarageQual"].value_counts().sort_values(ascending=False).plot.bar()
```

<br>
<br>

`TA is the most frequent value compare the sell price with Most frequent value and Null value.`

<br>
<br>

```python

sns.kdeplot(df[df["GarageQual"]=="TA"]["SalePrice"],label="Houese with TA")
sns.kdeplot(df[df["GarageQual"].isnull()]["SalePrice"],label="Houses with Nan")
plt.legend()
```

```python

temp = df[df["GarageQual"]=="TA"]["SalePrice"]
```

```python
df["GarageQual"] = df["GarageQual"].fillna('TA')
```

```python
df["GarageQual"].value_counts().sort_values(ascending=False).plot.bar()
```

```python

sns.kdeplot(temp,label="Houese with TA")
sns.kdeplot(df[df["GarageQual"]=="TA"]["SalePrice"],label="After filing Nan value with mode")
plt.legend()
```
