# 1. Scatter Plot: (Numerical - Numerical)

<br>

```python
tips.columns
```

```python

sns.scatterplot(x=tips['total_bill'],y=tips['tip']) # যারা বেশি টাকার খাবার খাচ্ছে তার বেশি টাকা টিপস দিচ্ছে । 
```

```python
# we want to know the either the customer was male or female?

sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

```python
# we can also see who is smoker:

sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

```python
# by that customer how many people came with him/her:

# we can also see who is smoker:
plt.figure(figsize=(8,8))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# Distplot (Numerical - Categorical)

<Br>
<Br>

```python
# instead of displot directly draw a kdeplot:

sns.kdeplot(x=titanic[titanic['survived'] == 0]['age'], label='Did not Survive',color='r',fill=True)
sns.kdeplot(x=titanic[titanic['survived'] == 1]['age'], label='Survived',color='y')
plt.legend()
```


`উপরের গ্রাফে কোন একটা বাচ্চার survived করার posibility বেশি dakhse । কারণ, হয়তো ঘটনার সময় বাচ্চাদের কে বাঁচানো জন্য অনেক বেশি effort দেওয়া হয়েছিলো । `
