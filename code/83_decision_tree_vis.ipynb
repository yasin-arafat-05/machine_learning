```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# using dtreeviz:
from dtreeviz import model
from IPython.display import Image,display_svg,SVG
```

```python

iris = load_iris()
iris.keys()
```

```python
X_train = iris.data
y_train = iris.target
```

```python
def tree(max_depth):
    classifier = DecisionTreeClassifier(max_depth=max_depth)
    classifier.fit(X_train,y_train)
    plot_tree(decision_tree=classifier)
```

```python

tree(None)
```

```python
tree(2)
```

```python

classifier = DecisionTreeClassifier(max_depth=3)
classifier.fit(X_train,y_train)
```

```python

"""
This may give us an error! 
We need to download graphviz in our pc.

For arch linux,
    sudo pacman -S graphviz
"""

viz = model(
    model=classifier,
    feature_names=iris.feature_names,
    X_train=X_train,
    y_train=y_train,
    class_names=iris.target_names
)

viz.view(scale=1.5)
```

# `# Regression Problem:`

<br>
<br>

```python

from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor
```

```python
diabetes = load_diabetes()
diabetes.keys()
```

```python

print(f"{diabetes.feature_names}")
print(f"{diabetes.data_filename}")
```

```python

X_train = diabetes.data
y_train = diabetes.target
```

```python

cls = DecisionTreeRegressor(criterion="poisson",max_depth=4)
cls.fit(X_train,y_train)
```

```python
plot_tree(cls)
```

```python

viz = model(
    model=cls,
    target_name='diabetes',
    feature_names=diabetes.feature_names,
    X_train=X_train,
    y_train=y_train
)

viz.view(scale=1.4)
```

```python

# change the orientation: 
viz.view(orientation="LR")
```
