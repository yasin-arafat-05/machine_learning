## a. Countplot: 

### we can see the count number:

` Hence, here we can treat the pclass,survived as a categorical data. like sex is a categorical data here either female or male. `

`Name is like categorical but here category is so many . we need  NLP to analysis this.`

<br>

```python

sns.countplot(x='survived',data=df)

df['survived'].value_counts()
```

```python
sns.countplot(x='pclass',data=df)
df['pclass'].value_counts()
```

```python
sns.countplot(x='sex',data=df)
df['sex'].value_counts()
```

```python

sns.countplot(x='embark_town',data=df)
df['embark_town'].value_counts()
```

<br>

# Pie Chart:

<br>

` Category exactly count the number of category but if we want to show it with percentage then we need piechart. `

<br>

```python
df['pclass'].value_counts().plot(kind="pie",autopct='%.2f')
```

```python
df['sex'].value_counts().plot(kind="pie",autopct='%.2f')
```
