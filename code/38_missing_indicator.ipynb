# Missing Indicator: 

<br>
<br>

```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

df = pd.read_csv("dataset/train.csv",usecols=["Age","Fare","Survived"])
df.head()
```

```python

X_train,X_test,y_train,y_test = train_test_split(
    df.drop(columns=["Survived"]),
    df["Survived"],
    test_size=0.2,
    random_state=2
)
```

<br>

# fill the value with simple imputer: 

<br>

```python

si = SimpleImputer(strategy="mean")
X_train_trans = si.fit_transform(X_train)
X_test_trans = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_trans,y_train)
y_pred = lr.predict(X_test_trans)
accuracy_score(y_test,y_pred) * 100
```

<br>
<br>
<br>

# Now, use Missing indicator: 

<br>
<br>
<br>

```python

mi = MissingIndicator()
X_train_miss = mi.fit_transform(X_train)
X_test_miss = mi.transform(X_test)
```

```python
X_train_miss
```

```python
X_train["Age_Na"] = X_train_miss
X_test["Age_Na"] = X_test_miss
```

```python

si = SimpleImputer(strategy="mean")
X_train_trans = si.fit_transform(X_train)
X_test_trans = si.transform(X_test)
```

```python

lr = LogisticRegression()
lr.fit(X_train_trans,y_train)
y_pred = lr.predict(X_test_trans)
accuracy_score(y_test,y_pred) * 100   # accuracy increase:
```


`In, SimpleImputer() we have the property for add Missing Indicator: `
`SimpleImputer(add_indicator=True)`
