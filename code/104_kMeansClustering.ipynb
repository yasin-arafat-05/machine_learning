# `# K Means Clustering:`

<br>
<br>

```python

import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.cluster import KMeans
```

```python

df = pd.read_csv("dataset/104_student_clustering.csv")
df.sample(2)
```

```python

sns.scatterplot(data=df,x='cgpa',y='iq')
```

```python

wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
wcss
```

```python
val = np.arange(start=1,stop=11,step=1)
val
```

```python


# Eblow Curve:
sns.lineplot(x=val,y=wcss)
```

```python

km = KMeans(n_clusters=4)
km.fit(X=df)
```

```python

# prediction: 
X = df.iloc[:,:]
predict = km.predict(X)
predict
```

```python

# convert dataFreame into numpy array
X = X.values
```

```python

# cgpa and iq value those who are in cluster: 1
print(X[predict==1])

# extract the iq value those who are in cluster: 1
print("*"*100)
print(X[predict==1,0])
```

```python

# draw the predicted cluster: 
sns.scatterplot(x=X[predict==0,0],y=X[predict==0,1],c="blue")
sns.scatterplot(x=X[predict==1,0],y=X[predict==1,1],c="red")
sns.scatterplot(x=X[predict==2,0],y=X[predict==2,1],c="black")
sns.scatterplot(x=X[predict==3,0],y=X[predict==3,1],c="yellow")
```

# `#K-means on 3D data:`

<br>
<br>

```python
from sklearn.datasets import make_blobs


centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=42)

```

```python

import plotly.express as px 
px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
```

```python


wcss= []

for i in range(1,12):
   km = KMeans(n_clusters=i,verbose=True) 
   km.fit(X=X)
   wcss.append(km.inertia_)
   
wcss
```

```python
val = np.arange(start=1,stop=12,step=1)
val
```

```python

sns.lineplot(x=val,y=wcss)
```

```python

km = KMeans(n_clusters=4)
km.fit_transform(X=X)
predict = km.predict(X=X)
predict
```

```python

df = pd.DataFrame()
df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = predict
```

```python
import warnings
warnings.filterwarnings("ignore")

px.scatter_3d(data_frame=df,x='col1',y='col2',z='col3',color='label')
```
