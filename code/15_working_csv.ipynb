# 12. Converter parameter:

<br>

`Some time we in a column if we want to repleace a data with another data. Like: in the IPL data set i want to change the name Royal Challergers Bangalore in RCB of all of the column`

<br>

```python
df = pd.read_csv("dataset/IPL Matches 2008-2020.csv")
df.head()
```

```python
def rename (name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name 
    
```

```python
df = pd.read_csv('dataset/IPL Matches 2008-2020.csv',converters={"team1":rename})
df.head() # now RCB here 
```

# 14. Loding a huge dataset in chunks:

<br>

`If our dataset is huge and if our ram can't road all the data  of that dataset then we can divide whole data set into smallest chunks. `

<br>

```python
df = pd.read_csv("dataset/aug_train.csv")
df.shape  #divide this dataset 
```

```python
dfs = pd.read_csv("dataset/aug_train.csv",chunksize=5000)
for chunks in dfs:
    print(chunks.shape)
    
```
